```python
from ultralytics import YOLO
import cv2
import os
from deep_sort_realtime.deepsort_tracker import DeepSort
from collections import defaultdict
import math
import locale
locale.getpreferredencoding = lambda: "UTF-8"
```

```python
drive_project_path = 'D:\\College\\Masters\\Laurier\\Winter2025\\GraduateProject\\YOLO_runs'
dataset_path = "D:\\College\\Masters\\Laurier\\badminton_Dataset\\Dataset_V1"
yaml_path = os.path.join(dataset_path, "data.yaml")
```

```python
# === Load & Train YOLO ===
model = YOLO("yolov8n.pt")

model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=0.80,
    project=drive_project_path,
    name="player_detection",
    save=True,
    save_period=1,
    cache=False,
    device=0
)

# Saving the best weights
best_weights = os.path.join(drive_project_path, "player_detection", "weights", "best.pt")
print(f" Best weights saved at: {best_weights}")
```
